```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace this with your own data)
data = np.random.rand(6, 5)  # A 5x5 array of random values between 0 and 1
print(data)
# Create a heatmap
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()  # Add a colorbar to show color intensity scale

# Optional: Customize axis labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Color Intensity Heatmap')

# Show the plot
plt.show()
```
